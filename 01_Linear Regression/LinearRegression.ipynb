##Linear Regression

Simple Linear Regression is a type of Regression algorithms that models the relationship between a dependent variable and a single independent variable. The relationship shown by a Simple Linear Regression model is linear or a sloped straight line, hence it is called Simple Linear Regression.

In this notebook we have implemented the linear regression without train test split and with train test split 

###Simple Linear Regression 

Step-1: Data Pre-processing

```python
# Import all the libraries 

import numpy as np 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
time_studied = np.array([20, 50, 32, 65, 23, 43, 10, 5, 22, 35, 29, 5, 56]).reshape(-1, 1)
scores = np.array([56, 83, 47, 93, 47, 82, 45, 78, 55, 67, 57, 4, 12]).reshape(-1, 1)

# x = scores.size
# x
```

Step-2: Fitting the Simple Linear Regression to the Training Set:

```python
model = LinearRegression()
# to find perfect line we use model.fit
model.fit(time_studied, scores)
# model.predict(np.array([56]).reshape(-1, 1))
```

Step: 3. Prediction of test set result & visualizing the Training set results:

```python
plt.scatter(time_studied, scores)
plt.plot(np.linspace(0, 70, 100).reshape(-1, 1), model.predict(np.linspace(0, 70, 100).reshape(-1, 1)), 'red')
plt.ylim(0, 100)
plt.xlabel('time_studied')
plt.ylabel('scores')
plt.grid(True)
plt.show
```

###Linear regression using Train Test Split

Step-1: Data Pre-processing

```python
import numpy as np
import matplotlib.pyplot as plot
from sklearn.model_selection import train_test_split
```

```python
time_studied = np.array([20, 50, 32, 65, 23, 43, 10, 5, 22, 35, 29, 5, 56]).reshape(-1, 1)
scores = np.array([56, 83, 47, 93, 47, 82, 45, 23, 55, 67, 57, 4, 89]).reshape(-1, 1)
# x = scores.size
# x
```

```python
# Splitting the dataset into training and test set.
time_train, time_test, score_train, score_test = train_test_split(time_studied, scores, test_size = 0.3)
```

Step-2: Fitting the Simple Linear Regression to the Training Set:

```python
model = LinearRegression()
model.fit(time_test, score_test)
    
print(model.score(time_train, score_train))
```

Step: 3. Prediction of test set result:

```python
model.predict(np.array([10]).reshape(-1, 1))
```

Step: 4. visualizing the Training set results:

```python
plt.scatter(time_train, score_train)
plt.plot(np.linspace(0, 70, 100).reshape(-1, 1), model.predict(np.linspace(0, 70, 100).reshape(-1, 1)),'red')
plt.show()
print(model.score(time_train, score_train))
```
